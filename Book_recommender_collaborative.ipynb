## **Collaborative Filtering based Recommender System**

```python
ratings.merge(books,on='ISBN')
```

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.shape
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_ratings.groupby('Book-Title').count()['Book-Rating']
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pt)
```

```python
cosine_similarity(pt).shape
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):
    # fetch index of the book
    index = np.where(pt.index== book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('1984')
```

```python
recommend('Message in a Bottle')
```
